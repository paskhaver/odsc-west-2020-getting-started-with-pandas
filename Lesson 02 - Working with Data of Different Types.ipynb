## Numeric Filtering

```python
nba["Salary"] > 35000000
high_earner = nba["Salary"] > 35000000
nba[high_earner]
```

```python
nba["Birthday"] < "1982-01-01"
is_older = nba["Birthday"] < "1982-01-01"
nba[is_older]
```

## Filtering with Multiple Conditions (AND + OR)

```python
nba["Position"] == "C"
is_center = nba["Position"] == "C"
```

```python
nba["Team"] == "Houston Rockets"
is_rocket = nba["Team"] == "Houston Rockets"
```

```python
nba[is_center & is_rocket]
```

```python
nba[is_center | is_rocket]
```

## The `isin` Method

```python
celtics = nba["Team"] == "Boston Celtics"
warriors = nba["Team"] == "Golden State Warriors"
hornets = nba["Team"] == "Charlotte Hornets"
```

```python
nba[celtics | warriors | hornets].head()
```

```python
nba["Team"].isin(["Boston Celtics", "Golden State Warriors", "Charlotte Hornets"])
target_teams = nba["Team"].isin(["Boston Celtics", "Golden State Warriors", "Charlotte Hornets"])
```

```python
nba[target_teams].head()
```

## The `between` Method

```python
one_mil_or_more = nba["Salary"] >= 1000000
two_mil_or_less = nba["Salary"] <= 2000000
nba[one_mil_or_more & two_mil_or_less].head()
```

```python
nba[nba["Salary"].between(1000000, 2000000)].head()
```

```python
nineties_kids = nba["Birthday"].between("1990-01-01", "1999-12-31")
nba[nineties_kids].head()
```
